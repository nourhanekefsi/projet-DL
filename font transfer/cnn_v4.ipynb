```python
import os
import tensorflow as tf
from tensorflow.keras import layers
from tensorflow.keras.optimizers import Adam
import matplotlib.pyplot as plt
import numpy as np

# Constants
DATASET_DIR = "/kaggle/input/transfer-font-style-texts/font_transfer_dataset/pairs"
MODEL_DIR = "/kaggle/working/cnn_model"
IMG_HEIGHT = 256
IMG_WIDTH = 256
BUFFER_SIZE = 10000
BATCH_SIZE = 32
EPOCHS = 50

# Function to load and preprocess images
def load_image(image_path):
    image = tf.io.read_file(image_path)
    image = tf.image.decode_png(image, channels=1)  # Grayscale
    image = tf.image.resize(image, [IMG_HEIGHT, IMG_WIDTH])
    image = image / 255.0  # Normalize to [0, 1]
    return image

def load_data(image_dir):
    input_paths = sorted([os.path.join(image_dir, f) for f in os.listdir(image_dir) if "input" in f])
    target_paths = sorted([os.path.join(image_dir, f.replace("input", "target")) for f in os.listdir(image_dir) if "input" in f])

    inputs = [load_image(p) for p in input_paths]
    targets = [load_image(p) for p in target_paths]
    dataset = tf.data.Dataset.from_tensor_slices((inputs, targets))
    return dataset

# Data augmentation
def data_augmentation(input_image, target_image):
    if tf.random.uniform(()) > 0.5:
        input_image = tf.image.flip_left_right(input_image)
        target_image = tf.image.flip_left_right(target_image)
    return input_image, target_image

# Build the CNN model (Encoder-Decoder)
def build_cnn_model():
    inputs = layers.Input(shape=(256, 256, 1))
    x1 = layers.Conv2D(64, (3, 3), padding="same", activation="relu")(inputs)
    x2 = layers.Conv2D(128, (3, 3), strides=2, padding="same", activation="relu")(x1)
    x3 = layers.Conv2D(256, (3, 3), strides=2, padding="same", activation="relu")(x2)
    
    # Upsample and concatenate with skip connections
    x4 = layers.Conv2DTranspose(128, (3, 3), strides=2, padding="same", activation="relu")(x3)
    x4 = layers.Concatenate()([x4, x2])  # Skip connection
    x5 = layers.Conv2DTranspose(64, (3, 3), strides=2, padding="same", activation="relu")(x4)
    x5 = layers.Concatenate()([x5, x1])  # Skip connection

    # Output layer
    outputs = layers.Conv2D(1, (3, 3), padding="same", activation="tanh")(x5)  # Tanh for output
    return tf.keras.Model(inputs, outputs)


# Display function for images
def display_generated_images(input_image, target, generated_image, epoch):
    plt.figure(figsize=(8, 8))
    images = [input_image, target, generated_image]
    titles = ['Input', 'Target', 'Generated']
    for i, (img, title) in enumerate(zip(images, titles)):
        plt.subplot(1, 3, i + 1)
        plt.imshow(img[0, :, :, 0], cmap='gray')
        plt.title(title)
        plt.axis('off')
    plt.suptitle(f'Epoch {epoch}')
    plt.show()

# Callback for displaying images at the end of each epoch
class DisplayCallback(tf.keras.callbacks.Callback):
    def __init__(self, dataset):
        super(DisplayCallback, self).__init__()
        self.dataset = dataset.take(1)

    def on_epoch_end(self, epoch, logs=None):
        for input_image, target in self.dataset:
            generated_image = self.model(input_image, training=False)
            display_generated_images(input_image.numpy(), target.numpy(), generated_image.numpy(), epoch + 1)

# Compile the model
model = build_cnn_model()
model.compile(optimizer=Adam(1e-4), loss="mean_squared_error", metrics=["accuracy"])

# Prepare the dataset
dataset = load_data(DATASET_DIR)
dataset = dataset.map(data_augmentation).shuffle(BUFFER_SIZE).batch(BATCH_SIZE)

# Train the model with the callback
display_callback = DisplayCallback(dataset)
model.fit(dataset, epochs=EPOCHS, callbacks=[display_callback])

# Save the model
model.save(MODEL_DIR)
```
